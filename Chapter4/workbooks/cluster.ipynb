The mean of a collection of nonnegative numbers is itself nonnegative (as sums and positive products of nonnegative numbers are nonnegative). 3.   If each $\boldsymbol{x}_i$ is a relative frequency vector, that is, each vector records the proportion of occurrences of certain categories, explain why each representative $\boldsymbol{z}_j$ is likewise a relative frequency vector.


````python
### EXERCISE 3#######################################
print('Answer here')
````

<details><summary>Click here for the solution</summary>

Relative frequency vectors are nonnegative, and we see that each $\boldsymbol{z}_j$ is likewise nonnegative. The sum of $\boldsymbol{z}_j$ will be the sum of averages of each entry from each vector $\boldsymbol{x}_i$ in its group. Using properties of summation, this is equivalent to the average of the vector sums. As each vector sums to 1, their average is likewise $1$. Thus, the sum of $\boldsymbol{z}_j$ is $1$. Therefore, $\boldsymbol{z}_j$ is a relative frequency vector.

</details>

6.   Suppose $k=2$. Then the function $f(\boldsymbol{x}) = \Vert \boldsymbol{x}-\boldsymbol{z}_2\Vert^2 - \Vert \boldsymbol{x}-\boldsymbol{z}_1\Vert^2$ can be used to determine which group $\boldsymbol{x}$ belongs to. If $f(\boldsymbol{x}) > 0$, then $\Vert \boldsymbol{x}-\boldsymbol{z}_2\Vert^2 - \Vert \boldsymbol{x}-\boldsymbol{z}_1\Vert^2>0$, which implies that $\Vert \boldsymbol{x}-\boldsymbol{z}_2\Vert^2 > \Vert \boldsymbol{x}-\boldsymbol{z}_1\Vert^2$ or more simply $\text{dist}(\boldsymbol{x}, \boldsymbol{z}_2) > \text{dist}(\boldsymbol{x}, \boldsymbol{z}_1)$. Thus, if $f(\boldsymbol{x})>0$, then $\boldsymbol{x}$ is closer to $\boldsymbol{z}_1$ and belongs to group $C_1$. Similarly, if $f(\boldsymbol{x})<0$, then $\boldsymbol{x}$ belongs to group $C_2$. Using properties of the inner product, expand and simplify the formula for $f$ by finding a vector $\boldsymbol{a}$ and scalar $b$ such that $f(\boldsymbol{x}) = \boldsymbol{a} \cdot \boldsymbol{x} + b$.

````python
### EXERCISE 6#######################################
print('Answer here')
````

<details><summary>Click here for the solution</summary>

$\begin{eqnarray*}
f(x) &=& \Vert \boldsymbol{x}-\boldsymbol{z}_2\Vert^2 - \Vert \boldsymbol{x}-\boldsymbol{z}_1\Vert^2 \\
&=& (\boldsymbol{x}-\boldsymbol{z}_2)\cdot (\boldsymbol{x}-\boldsymbol{z}_2) - (\boldsymbol{x}-\boldsymbol{z}_1)\cdot (\boldsymbol{x}-\boldsymbol{z}_1) \\
&=& (\boldsymbol{x}\cdot \boldsymbol{x} - \boldsymbol{x}\cdot \boldsymbol{z}_2 - \boldsymbol{z}_2\cdot \boldsymbol{x} + \boldsymbol{z}_2\cdot \boldsymbol{z}_2) - (\boldsymbol{x}\cdot \boldsymbol{x} - \boldsymbol{x}\cdot \boldsymbol{z}_1 - \boldsymbol{z}_1\cdot \boldsymbol{x} + \boldsymbol{z}_1\cdot \boldsymbol{z}_1) \\
&=& (\boldsymbol{x}\cdot \boldsymbol{x} - \boldsymbol{x}\cdot \boldsymbol{x}) - (2\boldsymbol{z}_2)\cdot \boldsymbol{x} + (2\boldsymbol{z}_1)\cdot \boldsymbol{x} + (\boldsymbol{z}_2\cdot \boldsymbol{z}_2 - \boldsymbol{z}_1\cdot \boldsymbol{z}_1) \\
&=& 2(\boldsymbol{z}_1-\boldsymbol{z}_2)\cdot \boldsymbol{x} + (\boldsymbol{z}_2\cdot \boldsymbol{z}_2 - \boldsymbol{z}_1\cdot \boldsymbol{z}_1)
\end{eqnarray*}$

Therefore, let $\boldsymbol{a}= 2(\boldsymbol{z}_1-\boldsymbol{z}_2)$ and $b = \boldsymbol{z}_2\cdot \boldsymbol{z}_2 - \boldsymbol{z}_1\cdot \boldsymbol{z}_1 = \Vert\boldsymbol{z}_2\Vert^2 - \Vert \boldsymbol{z}_1\Vert^2$.

</details>
